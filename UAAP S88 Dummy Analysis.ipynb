```python
import pandas as pd

data = {
    'team': ['NU', 'UP', 'UST', 'DLSU', 'ADMU', 'ADU', 'FEU', 'UE'],
    'wins': [11, 9, 7, 6, 5, 6, 5, 0],
    'losses': [2, 3, 5, 6, 7, 7, 7, 13],
    'games_played': [13, 12, 12, 12, 12, 13, 12, 13]
}

df = pd.DataFrame(data, columns=['team', 'wins', 'losses', 'games_played'])
df['win_percentage'] = (df['wins'] / df['games_played']*100).round(2)
df['remaining_games'] = 14 - df['games_played']
df.sort_values(by='win_percentage', ascending=False, inplace=True)
display(df)
```

```python
#Top 4 teams as of now
top4 = df[0:4]
display(top4)

top_team = df['wins'].idxmax()
```

```python
print(f'The average win rate among all UAAP teams is {df['win_percentage'].mean().round(0)}%')
```
